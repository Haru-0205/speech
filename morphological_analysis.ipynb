```python
f = open('lyrics.txt', 'r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
f.close
```

```python
print(data.splitlines)
```

```python
from janome.tokenizer import Tokenizer
```

```python
t = Tokenizer()
```

```python
for token in t.tokenize(data):
    print(token)
```

```python
lyrics = data.replace("\n", "")
```

```python
lyrics = lyrics.replace(" ","")
```

```python
print(lyrics)
```

```python
import re
words=[]

for token in t.tokenize(lyrics):
    if re.compile("^助詞").search(token.part_of_speech) or re.compile("^助動詞").search(token.part_of_speech) :
        print("")
    else:
        words.append(token.base_form)
        print(token)
```

```python
print(words)
```

```python
import collections

count_words = collections.Counter(words)
```

```python
print(count_words)
```

```python
from wordcloud import WordCloud

fpath = 'C:/Windows/Fonts/YuGothM.ttc'
text = ' '.join(words)

wordcloud = WordCloud(background_color='white', font_path=fpath, width=600, height=450).generate(text)
wordcloud.to_file('./wordcloud-i_wanna_be_a_girl.png')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["font.family"] = "Yu Gothic"
plt.bar(list(count_words.keys()), list(count_words.values()))
```

```python
for_graph = count_words.most_common(10)
print(for_graph)
```
